# Gaussian distribution
## numpy.random.Generator.normal()

```python
mu, sigma = 0, 0.1 # Mean and standard deviation
s = rng.normal(mu, sigma, 10000)
print(np.mean(s))
print(np.std(s, ddof=1))
```

```python
import matplotlib.pyplot as plt
count, x, ignored = plt.hist(s, 50, density=True)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')
plt.show()
```

# Add Gaussian noise
- get height and width of sample image

```python
height, width, tmp = img.shape
```

```python
img_noise = np.zeros((height, width), dtype=float)
```

- create gaussian noise

```python
mean = 0
stddev = 0.5 ** 0.5

noise = np.random.normal(mean, stddev, img.shape)
plt.imshow(noise)
plt.show()
```

- add gaussian noise to image

```python
import cv2

noisy_image = cv2.add(img, noise.astype(np.uint8))
```
